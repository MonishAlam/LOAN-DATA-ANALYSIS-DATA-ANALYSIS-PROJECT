```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df=pd.read_csv('LOAN_ANALYSIS_NEW_DATA.CSV')
```

```python
print(df)
```

```python
pd.isnull(df).sum()
```

```python
ax=sns.countplot(x='Gender',data=df,hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
plt.title('Gender distribution of loan applicants')
```

The majority of bank loan recepients are male

```python
sns.countplot(x='Gender',data=df,hue='Married')
plt.title('Count of loan applicants by gender and marital status')
```

Male loan recipients are mostly married, whereas females are predominantly unmarried

```python
sns.countplot(x='Education',data=df,hue='Education')
plt.title('Counnt of loan applicants by education')
```

The majority of loan recipients are graduates

```python
sns.countplot(x='Self_Employed',data=df,hue='Self_Employed')
plt.title('Count of loan applicants by self employment status')
```

Most loan applicants are not self employed

```python
sns.histplot(x='ApplicantIncome',kde=True,label='Applicants income',data=df)
sns.histplot(x='CoapplicantIncome',kde=True,label='Coapplicant income',data=df)
plt.title('Income distribution of applicant vs coapplicant')
plt.legend()
```

Most coapplicants have zero income,and the mojority of applicants earn between 2000 to 10000, with few exceeding 20000.

```python
approval_percentage=df['Loan_Status'].value_counts(normalize=True)['Y']*100
ax=plt.bar(['Approved','Not Approved'],[approval_percentage,100-approval_percentage])
plt.title('Percentage of people with approved loans')
plt.xlabel('Loan approval status')
plt.ylabel('percentage');    # by using ; the text above graph does not display \
```

The mojority of individuals have their loans approved 

```python
sns.scatterplot(x='ApplicantIncome',y='LoanAmount',data=df)
plt.title('Relationship between applicants income and loan amount')
sns.set(rc={'figure.figsize':(14,5)})
```

people with less money usually get smaller loans, even those with a bit more money do the same.

```python
sns.scatterplot(x='CoapplicantIncome',y='LoanAmount',data=df)
plt.title('Impact of co applicant income on loan amount')
plt.xlabel('coappliants income')
plt.ylabel('loan amount')
```

Lower coapplicants incomes usually mean smaller loans, with few exceptions for larger amounts.

```python
ax=sns.countplot(x='Property_Area',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
plt.title('Distribution of loan applicants property area')
```

Loan applicants are predominantly spread across different residential areas, with notable presence in semiurban areas followed by urban and rural.

```python
sns.countplot(x='Credit_History',data=df,hue='Loan_Status')
plt.title('influence of credit history on loan status')
```

Applicants with credit history 0 mostly get loan status "NO"
Applicants with credit history 1 mostly get loan status "YES", but with some exceptions of "NO"

CONCLUSION : The majority of bank loan applicants are graduated married male who earn between 2000 and 10000 who have their loans approved and have lesser income, usually get smaller loans among which majority of them reside in semiurban area.
majority of them have credit history of "1" and have their loan status "yes".
